```python
import pickle
import pandas as pd
```

```python
# read r2gengen.pickle
with open('r2gengen.pickle', 'rb') as f:
    r2gengen = pickle.load(f)
```

```python
r2gengen
```

```python
import pandas as pd
import numpy as np

# Your list of dictionaries

# Convert the list of dictionaries into a DataFrame
df_list = [pd.DataFrame(d) for d in r2gengen]

# Concatenate all dataframes along the vertical axis
df = pd.concat(df_list)

# Calculate mean and standard deviation
df_mean = df.groupby(df.index).mean()
df_std = df.groupby(df.index).std()

# Rename the columns
df_mean.columns = [f'{col}_mean' for col in df_mean.columns]
df_std.columns = [f'{col}_std' for col in df_std.columns]

# Join the two dataframes
df_result = df_mean.join(df_std)

print(df_result)
```
